```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from google.colab import files

# Función para codificar la operación en 2 bits
def encode_operation(op_str):
    # suma=00, resta=01, mult=10, div=11
    mapping = {
        'sum': [0,0],
        'sub': [0,1],
        'mul': [1,0],
        'div': [1,1]
    }
    return mapping[op_str]

# Generar datos de entrenamiento
def generate_data(num_samples=10000):
    X = []
    y = []
    operations = ['sum', 'sub', 'mul', 'div']

    for _ in range(num_samples):
        x1 = np.random.uniform(1, 10)
        x2 = np.random.uniform(1, 10)
        op = np.random.choice(operations)

        op_bits = encode_operation(op)

        X.append([x1, x2, op_bits[0], op_bits[1]])

        if op == 'sum':
            y.append(x1 + x2)
        elif op == 'sub':
            y.append(x1 - x2)
        elif op == 'mul':
            y.append(x1 * x2)
        elif op == 'div':
            y.append(x1 / x2)

    return np.array(X), np.array(y)

# Crear y compilar modelo
model = Sequential()
model.add(Dense(20, input_dim=4, activation='relu'))
model.add(Dense(1))  # Salida: resultado numérico

model.compile(optimizer=Adam(learning_rate=0.01), loss='mse')

# Generar datos
X_train, y_train = generate_data(20000)

# Entrenar la red
model.fit(X_train, y_train, epochs=30, batch_size=64, verbose=2)

# Prueba ejemplo
def predict_operation(x1, x2, op_str):
    op_bits = encode_operation(op_str)
    input_data = np.array([[x1, x2, op_bits[0], op_bits[1]]])
    pred = model.predict(input_data)
    print(f"{op_str}({x1:.2f}, {x2:.2f}) ≈ {int(round(pred[0,0]))}")


predict_operation(4, 3, 'sum')
predict_operation(4, 3, 'sub')
predict_operation(2, 3, 'mul')
predict_operation(2, 2, 'div')
model.save("modelo_operaciones.h5")
```
